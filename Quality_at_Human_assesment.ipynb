```python
modelList = {'skipgram_500_10cw_10mc_fastext':skipgram_500_10cw_10mc_fastext,
             'skipgram_500_10cw_5mc_fastext':skipgram_500_10cw_5mc_fastext,
             'skipgram_500_10cw_3mc_fastext':skipgram_500_10cw_3mc_fastext,
             'skipgram_500_10cw_2mc_fastext':skipgram_500_10cw_2mc_fastext,
             
             'skipgram_500_5cw_10mc_fastext':skipgram_500_5cw_10mc_fastext,
             'skipgram_500_5cw_5mc_fastext':skipgram_500_5cw_5mc_fastext,
             'skipgram_500_5cw_3mc_fastext':skipgram_500_5cw_3mc_fastext,
             'skipgram_500_5cw_2mc_fastext':skipgram_500_5cw_2mc_fastext,
             
             'skipgram_500_7cw_10mc_fastext':skipgram_500_7cw_10mc_fastext,
             'skipgram_500_7cw_5mc_fastext':skipgram_500_7cw_5mc_fastext,
             'skipgram_500_7cw_3mc_fastext':skipgram_500_7cw_3mc_fastext,
             'skipgram_500_7cw_2mc_fastext':skipgram_500_7cw_2mc_fastext
            }
```

```python
def get_score(table,model,src,trgt):
    """
    Get the similarity score between the concept present in the table if they are present in the the model
    src and trgt are then column number for which similarity score has be calculated.
    """
    similar_score = []
    for i,rows in table.iterrows():
        src_n = True if rows[src] in model.wv.vocab else False
        trgt_n = True if rows[trgt] in model.wv.vocab else False

        if src_n  and trgt_n :
            #print('[{}]: {}, {}, {}'.format(i,rows[2],rows[3],model.similarity(rows[2],rows[3])))
            similar_score.append(model.wv.similarity(rows[src],rows[trgt]))
        elif src_n:
            #print('[{}]: target Concept {} not present in vocab of model'.format(i,rows[trgt]))
            similar_score.append(0)
        elif trgt_n :
            similar_score.append(0)
            #print('[{}]: source concept {} not present in vocab of model'.format(i,rows[src]))
                
        else:
            similar_score.append(0)
            #print('[{}]: {}, {} are not present in vocab of model'.format(i,rows[src],rows[trgt]))
    #print(skipgram_500_10cw_2mc_fastext.similarity(rows[2],rows[3]))
    return similar_score
```

```python
for k,model in modelList.items():
    for i,row in mini.iterrows():
        #print(row[6],row[7])
        mini[k] = get_score(mini,model,2,3)
```
